```python
import numpy as np
import matplotlib.pyplot as plt
# mnist_cnn
epoch_mnist = np.arange(12)+1
acc_mnist = np.array([0.9159,0.9724,0.9790,0.9826,0.9850,0.9870,0.9878,0.9890,0.9896,0.9908,0.9911,0.9918])
loss_mnist = np.array([0.2727,0.0929,0.0709,0.0578,0.0475,0.0433,0.0393,0.0365,0.0336,0.0311,0.0295,0.0282])
fig1 = plt.figure(1)
ax1_1 = fig1.add_subplot(1,2,1)
ax1_1.plot(epoch_mnist,acc_mnist)
ax1_1.set_xlabel('epochs')
ax1_1.set_ylabel('accuracy')
ax1_2 = fig1.add_subplot(1,2,2)
ax1_2.plot(epoch_mnist,loss_mnist)
ax1_2.set_xlabel('epochs')
ax1_2.set_ylabel('loss')
fig1.tight_layout()
fig1.savefig('mnist_results.png')
fig1.show()
```

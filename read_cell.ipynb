```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt

parapath = './cell_images/Parasitized/'
uninpath = './cell_images/Uninfected/'
paralist = os.listdir(parapath)
uninlist = os.listdir(uninpath)
cell=[]
label=[]
for i in paralist:
    try:
        image =  Image.open(parapath+i).resize((32,32))
        data = np.array(image)/255.
        cell.append(data)
        label.append(1)
    except:
        print(i+' is not a image file, skipping it')
for j in uninlist:
    try:
        image =  Image.open(uninpath+j).resize((32,32))
        data = np.array(image)/255.
        cell.append(data)
        label.append(0)
    except:
        print(j+' is not a imgae file, skipping it')
cell=np.array(cell)
label=np.array(label)
#np.save('cell',cell)
#np.save('label',label)
```

```python
cell=np.load('cell.npy')
label=np.load('label.npy')
print('(number; size*size; rgb) = ', cell.shape)
fig1=plt.figure(1)
ax1_1=fig1.add_subplot(1,2,1)
ax1_1.imshow(cell[0])
ax1_1.set_title('parasitized cell, label=1')
ax1_2=fig1.add_subplot(1,2,2)
ax1_2.imshow(cell[-1])
ax1_2.set_title('uninfected cell, label=0')
fig1.suptitle('examples')
fig1.tight_layout()
fig1.show()
```
